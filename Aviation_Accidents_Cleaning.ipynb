## Data Cleaning

### Filtering aircrafts and events

We want to filter the dataset to include aircraft that the client is interested in an analysis of:
- inspect relevant columns
- figure out any reasonable imputations
- filter the dataset

```python
#Make a copy
airline_df = df.copy(deep=True)
```

```python
#Remove whitespaces and make lowercase letter 
airline_df.columns = airline_df.columns.str.replace(".", "_", regex=False).str.lower().str.strip()
```

Doing this makes all column names uniform

```python
#Change event date and publication date to datetime datatype
airline_df['publication_date']= pd.to_datetime(airline_df['publication_date'], format= '%d-%m-%Y')
airline_df['event_date']= pd.to_datetime(airline_df['event_date'], format= '%Y-%m-%d')

#Filter event date from 1983 onwards
airline_df = airline_df[airline_df['event_date'].dt.year >= 1983]
```

Changed the data type of eventdate and publication date to datetime and filtered event date to 1983 onwards

```python
#Filter column inestigation type to have only rows with accident
airline_df= airline_df[airline_df['investigation_type'] == 'Accident']

#Filter amateur aircrafts to have only professional build
airline_df= airline_df[airline_df['amateur_built'] == 'No']
```

Filtering investigation type to accident only and amateur build to 'no' as specified for analysis

### Cleaning and constructing Key Measurables

Injuries and robustness to destruction are a key interest point for the client. Clean and impute relevant columns and then create derived fields that best quantifies what the client wishes to track. **Use commenting or markdown to explain any cleaning assumptions as well as any derived columns you create.**

**Construct metric for fatal/serious injuries**

*Hint:* Estimate the total number of passengers on each flight. The likelihood of serious / fatal injury can be estimated as a fraction from this.

```python
#Filling in null values in numerical columns of injuries
airline_df= airline_df.fillna({'total_fatal_injuries':0,
                                'total_serious_injuries':0, 
                                'total_minor_injuries':0, 
                                'total_uninjured':0})
```

Observation: Filling injury and people whosurvived with 0 because its the mode for all columns and doing this so that calculations can be done without null


```python
#Calculating total passengers onboard
airline_df['total_passengers'] = (airline_df['total_fatal_injuries'] + 
                                  airline_df['total_serious_injuries'] + 
                                  airline_df['total_minor_injuries'] +
                                   airline_df['total_uninjured'])
```

Observation: Ive created a column of total passengers

```python
#Fraction of serious and fatal injuries on each flight
airline_df['serious_and_fatal_fraction'] = (airline_df['total_fatal_injuries'] + airline_df['total_serious_injuries']) / airline_df['total_passengers']
```

Observation: Ive created column of fraction of fatal and serios injuries 

**Aircraft.Damage**
- identify and execute any cleaning tasks
- construct a derived column tracking whether an aircraft was destroyed or not.

```python
#Before deriving column, impute the null vales with mode
airline_df['aircraft_damage'] = airline_df['aircraft_damage'].fillna(airline_df['aircraft_damage'].mode()[0])

#Create column
airline_df['aircraft_destroyed'] = airline_df['aircraft_damage'] == 'Destroyed'
```

Observation: Ive created a column with bool values to tell whther or not the aircraft was destroyed or not

### Investigate the *Make* column
- Identify cleaning tasks here
- List cleaning tasks clearly in markdown
- Execute the cleaning tasks
- For your analysis, keep Makes with a reasonable number (you can put the threshold at 50 though lower could work as well)

```python
#Identifying what should be cleaned in make column
unique_value = airline_df['make'].unique()

print(f'{'make'}\n {unique_value}\n')

print(f'Missing values : {airline_df['make'].isna().sum()}')
```

List of cleaning: 
1. Theres inconsistent casing of values
2. Same names appear in different spellings
3. There are 20 missing values

```python
#Cleaning the make column
airline_df['make'] = airline_df['make'].str.strip().str.capitalize()

#Drop the null values 
airline_df = airline_df.dropna(subset=['make'])

#Replace the names with multiple spellings as one
replacements = {
    'Piper aircraft': 'Piper',
    'Beech': 'Beechcraft',
    'Boeing co': 'Boeing',
    'Mcdonnell douglas aircraft co': 'Mcdonnell douglas',
    'Grumman american': 'Grumman',
    'Citrus design corp': 'Citrus',
    'Robinson helicopter company': 'Robinson helicopter',
    'Air tractor inc':'Air tractor'
    }

airline_df['make'] = airline_df['make'].replace(replacements)
```

```python
# Keep make threshold to 50
make_count = airline_df['make'].value_counts()
threshold_count = make_count[make_count >= 50].index
airline_df = airline_df[airline_df['make'].isin(threshold_count)]
```

Observation: Ive made the column make to have only incidents above 50
